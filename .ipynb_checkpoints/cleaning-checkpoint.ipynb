```python
import pandas as pd
characters = pd.read_csv("data/character-predictions_pose.csv")
battles = pd.read_csv("data/battles.csv")
```

```python
characters.head()
```

```python
subset = ["attacker_king","defender_king","attacker_1","attacker_2","attacker_3","attacker_4","defender_1","defender_2","defender_3","defender_4","attacker_commander","defender_commander","note"]
battles_with_names = battles[subset].fillna("")
def in_battle(name, row):
    count = 0
    for i in range(len(battles)):
        if any(name in battles_with_names.iloc[row].iloc[i] for i in range(len(subset))):
            count+=1
    return count
battle_counts = characters['name'].apply(lambda x: in_battle(x, 0))
```

```python
characters.loc[battle_counts > 0]
```
